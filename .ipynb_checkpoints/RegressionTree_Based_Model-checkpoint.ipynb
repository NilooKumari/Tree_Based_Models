## Decision tree : overview 
    - Can solve both regression and classification problems
    - Classic use of decision tree is analysis of segments in business data 
    - Decision tree classifiers recursively subset data to revel interesting patterns
    - Data need to be split in such a way that subsets of data end up being dominated by one class of the target variable

## Decision tree : algorithm
    - Good split in context of classification problem: Greater the class imbalance,better the split
    - Class imbalance can be measured by computing Gini or Entropy
    - For each split the purity metric is computed
    - Choose the variable which results in lowest value of purity metric
    - Continue doing these till  some stopping criteria is met 
    
- Stopping Criterion
    - Depth of the tree
    - Improvement in purity metrics
    - Value in terminal node

## Decision tree : performance metrics
    - Confusion metrics
    - ROC curve
    - Area under the curve
- At the model level of decision tree rules are decided for predicting probabilities and classes

## Decision tree : parameters and hyperparameters
    - Purity metrics: Gini or Entropy
    - Depth of the tree
- These parameters are estimated using cross validation

```python
# Standard imports

import os 
import pandas as pd

os.chdir("C:\\Users\\UJJWAL\\Desktop\\PythonProgramming\\Tree_Based_Models")
```

```python
# Import the data 
data= pd.read_csv("dm.csv")
```

```python
    data.head() # Data of e-retailer: demographic information of customers and their quarterly mean expenditure
```

```python
# Store feature matrix in "X"
X= data.drop(['Cust_Id', 'AmountSpent'],axis=1)
```

```python
# Convetring non-numeric variable to numeric in feature matrix
X=pd.get_dummies(X)
X.head()
```

```python
# Store response vector in "y"
y=data['AmountSpent']
```

```python
# Step 1: split X and y into training and testing sets
import sklearn.model_selection as model_selection
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.2,random_state=200)
```

```python
# Print the shapes of the new X objects
print (X_train.shape)
print (X_test.shape)
```

```python
# Print the shapes of the new y objects
print (y_train.shape)
print (y_test.shape)
```

```python
# Step 2: Train the model on the training data
import sklearn.tree as tree
reg=tree.DecisionTreeRegressor(max_depth=3)
reg.fit(X_train,y_train)
```

```python
# Accuracy score on the test data set
reg.score(X_test,y_test) # prints the mean squared error
```

```python
# Print which predictors are important
reg.feature_importances_
```

```python
pd.Series(reg.feature_importances_, index=X.columns).sort_values(ascending=False).head(5)
# Based on below result 'Salary' is the most important feature followed by 'Catalogs' and so on
```
